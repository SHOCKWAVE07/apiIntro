```python
import pandas as pd
import requests 
```

```python
response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=698340fe4c89972ca6d4420ecfc3c8a3&&language=en-US&page=1')
```

```python
response.json()['results']
```

```python
temp_df=pd.DataFrame(response.json()['results'])
temp_df.head()
```

```python
temp_df.shape
```

```python
temp_df = temp_df[['id','title','overview','release_date','popularity','vote_average','vote_count']]
temp_df.head()
```

```python
df=pd.DataFrame()
for i in range(1,500):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=698340fe4c89972ca6d4420ecfc3c8a3&&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df=pd.concat([df,temp_df],ignore_index=True)
   
df.head(25)
```
